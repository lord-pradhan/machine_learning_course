```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import InputLayer, Conv2D, MaxPooling2D
from keras import backend as K
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import model_selection

from keras.utils import np_utils
from keras.layers.core import Dense, Activation, Dropout
```

```python
# data preprocessing
raw_data = pd.read_csv('data.csv', skiprows=0)
# raw_data = np.genfromtxt('data.csv', delimiter=',' ,dtype=None)

temp = np.arange(0, raw_data.shape[1], 2).tolist()
features = np.zeros( (int(raw_data.shape[0]*raw_data.shape[1]/2) , 2))
labels = np.zeros((features.shape[0], 1))

count=0
for i in temp:
    features[count*raw_data.shape[0] : (count+1)*raw_data.shape[0] , :] = raw_data.values[:, i:i+2]
    labels[count*raw_data.shape[0] : (count+1)*raw_data.shape[0] , 0] = count
    count=count+1
data_comb = np.append(features,labels, axis=1)
data_train, data_test = sk.model_selection.train_test_split( data_comb, test_size=0.2, train_size= 0.8 )
print(data_train.shape, data_test.shape)

X_train_pure = data_train[:, 0:data_train.shape[1]-1]
Y_train_pure = data_train[:, data_train.shape[1]-1]

X_test_pure = data_test[:, 0:data_test.shape[1]-1]
Y_test_pure = data_test[:, data_test.shape[1]-1]

# the classes vary from 0 to 9
num_classes = 20

# convert class vectors to binary class matrices: one hot vector
Y_train = keras.utils.to_categorical(Y_train_pure, num_classes)
Y_test = keras.utils.to_categorical(Y_test_pure, num_classes)
print('y_train shape:', Y_train.shape)
print('y_test shape:', Y_test.shape)

# pre-processing: divide by max and substract mean
scale = np.max(X_train_pure)
X_train = X_train_pure/ scale
X_test = X_test_pure/ scale

mean = np.std(X_train)
X_train -= mean
X_test -= mean

input_dim = X_train.shape[1]
nb_classes = Y_train.shape[1]

# Build layers
model = Sequential()
model.add(Dense(128, input_dim=input_dim))
model.add(Activation('relu'))

model.add(Dense(64))
model.add(Activation('relu'))

model.add(Dense(nb_classes))
model.add(Activation('softmax'))

# we'll use categorical xent for the loss, and RMSprop as the optimizer
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adadelta())

print("Training...")
model.fit(X_train, Y_train, nb_epoch=250, batch_size=16, validation_split=0.1,  verbose=2)

print("Generating test predictions...")
preds = model.predict_classes(X_test, verbose=0)


# Compute test accuracy
count = 0
for i in range(Y_test.shape[0]):
    count += int(preds[i] == Y_test_pure[i])
test_acc = float(count)/Y_test.shape[0]
```

```python
print('Test accuracy is ', test_acc*100, '%')
```
